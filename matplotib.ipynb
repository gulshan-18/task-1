```python
#www.matplotlib.org
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#functional method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel("Y Label")
plt.title("Title")
```

```python
#used to plot multiple plots on a single figure
#plt.sublot(no. of rows,no. of columns ,position)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'b*-')
```

# Matplotlib Object Oriented Method.

```python
#create a figure
fig=plt.figure()
#add axis to set of figure
axes=fig.add_axes([0.1,0.1,0.8,0.8])#left, bottom, width, height (range 0 to 1)
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title("Larger plot")
axes2.plot(y,x)
axes2.set_title("Smaller plot")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
# Axes is an array of axes to plot on it is like an array
#axes.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')

#to overcome overlap problem between figures
plt.tight_layout()
```

# Figure Size and DPI

```python
fig=plt.figure(figsize=(8,2),dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_ylabel("Y")
ax.set_xlabel("X")
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend(loc=10)
#loc(0.1,0.1) loc works for where the label will be there
```
